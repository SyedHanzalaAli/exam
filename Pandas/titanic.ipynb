```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.columns
```

```python
df.info()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df
```

```python
df.info()
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.info()
```

```python
overAllSurvivalRate = df["Survived"].mean() * 100
print(overAllSurvivalRate)
```

```python
survival = df[df["Survived"] == 1]
survivalCount = survival["Survived"].value_counts()
print(survivalCount)
```

```python
total = len(df)
print(total)
```

```python
survival = df[df["Survived"] == 0]
survivalCount = survival["Survived"].value_counts()
print(survivalCount)
```

```python
genderSurvival = df.groupby("Sex")["Survived"].mean() * 100
print(genderSurvival)
```

```python
classSurvival = df.groupby("Pclass")["Survived"].mean() * 100
print(classSurvival)
```

```python
classGenderSurvival = df.groupby(["Pclass","Sex"])["Survived"].mean() * 100
print(classGenderSurvival)
```

```python
ageBins = [0,16,25,50,100]
ageLabels = ["Child" , "Young Adult" , "Adult" , "Senior"]

df["AgeGroup"] = pd.cut(df["Age"] , bins=ageBins , labels=ageLabels , right=True)
```

```python
df
```

```python
AgeGroupSurvival = df.groupby("AgeGroup" , observed=True)["Survived"].mean() *100
print(AgeGroupSurvival)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(8,5))

survivedbyClass = df.groupby("Pclass")["Survived"].sum()

ax.bar(survivedbyClass.index , survivedbyClass.values , color = ["orange" ,"green" ,"blue"])

ax.set_title("total survivors by passenger class")
ax.set_xlabel("Passenger class")
ax.set_ylabel("No of survivors")
ax.set_xticks([1,2,3])
```

```python
fig, ax = plt.subplots(figsize=(8,5))

survivedbyClass = df.groupby("Pclass")["Survived"].sum()

ax.barh(survivedbyClass.index , survivedbyClass.values , color = ["orange" ,"green" ,"blue"])

ax.set_title("total survivors by passenger class")
ax.set_xlabel("no of survivors")
ax.set_ylabel("Passenger class")
ax.set_yticks([1,2,3])
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.hist(df["Age"] , edgecolor = "black")

ax.set_title("Passenger Ages")
ax.set_xlabel("Age")
ax.set_ylabel("Frequency")
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.hist(df["Age"] ,bins=4, edgecolor = "black")

ax.set_title("Passenger Ages")
ax.set_xlabel("Age")
ax.set_ylabel("Frequency")
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.hist(df[df["Survived"]==1]["Age"] ,bins=20, edgecolor = "black")

ax.set_title("Passenger survivals")
ax.set_xlabel("Age")
ax.set_ylabel("Frequency")
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.hist(df[df["Survived"]==0]["Age"] ,bins=20, edgecolor = "black")

ax.set_title("Passenger survivals")
ax.set_xlabel("Age")
ax.set_ylabel("Frequency")
```

```python
fig, ax = plt.subplots(1,2 ,figsize=(20,5))

ax[0].hist(df[(df["Survived"]==1) & (df["Sex"]=="male")]["Age"],bins=20,edgecolor = "black")

ax[0].set_title("Passenger survivals male")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Frequency")

ax[1].hist(df[(df["Survived"]==1) & (df["Sex"]=="female")]["Age"],bins=20,edgecolor = "black" ,color = "pink")

ax[1].set_title("Passenger survivals female")
ax[1].set_xlabel("Age")
ax[1].set_ylabel("Frequency")
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ageCounts = df["Age"].round().value_counts().sort_index()

ax.plot(ageCounts.index, ageCounts.values)

ax.set_title("Passenger counts")
ax.set_xlabel("Age")
ax.set_ylabel("number")
```
