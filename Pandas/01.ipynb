```python
data = {"Book title" : ["The ALchemist" ,"Forty rules of love","Sapiens" , "Atomic Habit"],
        "Author" : ["Paulo Coelho", "Elif Shafak" , "Yuval Noah Harari" , "James Clear"],
        "Price PKR" : [1200,1500,2200,1800],
        "Stock" : [15,25,10,30]}
```

```python
def checkStock(bookTitle):

    bookData = df[df["Book title"] == bookTitle]

    if not bookData.empty:
        stockCount = bookData["Stock"].iloc[0]
        print(f"Stock for {bookTitle} : {stockCount}")
    else:
        print(f"Book not found : {bookTitle}")

checkStock("Sapiens")
checkStock("The Hobbit")
```
